```python
ohe = OneHotEncoder(drop='first')
```

```python
ohe_state = cust_df[['state']]
```

```python
ohe.fit(ohe_state)
```

```python
state_encoded = ohe.transform(ohe_state)
state_encoded
```

```python
state_encoded.todense()
```

```python
dummy_df = pd.DataFrame(state_encoded.todense(), columns=ohe.get_feature_names())
dummy_df
```

```python
cust_df = pd.concat([cust_df, dummy_df], axis=1)
cust_df
```
